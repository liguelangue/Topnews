```python
"""
关键词提取
"""
import jieba
import jieba.analyse as analyse
import pandas as pd
import codecs
```

```python
df = pd.read_csv("./data/news.csv", usecols=['zh_titles'], encoding='utf-8')
# 文件路径必须是英文，否则pandas读取会有问题
# df.tail(10)
```

```python
'''
中文关键词提取 jieba TF-IDF
jieba.analyse.extract_tags(sentence, topK=20, withWeight=False, allowPOS=())
sentence 为待提取的文本
topK 为返回几个 TF/IDF 权重最大的关键词，默认值为 20
withWeight 为是否一并返回关键词权重值，默认值为 False
allowPOS 仅包括指定词性的词，默认值为空
'''
def TF_IDF(sentence):
    seg_list = jieba.cut(sentence, cut_all=True)
    print(".. ".join(seg_list))
    keywords = jieba.analyse.extract_tags(sentence, topK=20, withWeight=False, allowPOS=('n','nr','ns'))
    print(keywords)
    content_S.append(keywords)
    return keywords
```

```python
import gensim
from gensim import corpora, models, similarities, matutils
import numpy as np
```

```python
#做映射，相当于词袋
dictionary = corpora.Dictionary(content_S)#做成字典
corpus = [dictionary.doc2bow(sentence) for sentence in content_S]#在每个句子中都做自然模型
```

```python
#建立LDA模型，传入语料、映射字典、指定要得到多少个主题(无监督)
lda = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=20) #类似Kmeans自己指定K值
```

```python
f = codecs.open("./data/topics.txt", "w", "utf-8")    # 打开文件以便写入中文
#分别打印每个主题各个最具代表性的5个关键词
for topic in lda.print_topics(num_topics=20, num_words=5):
    print (topic[1], file = f)
f.close()  #  关闭文件
# print(len(corpus))
```

```python
"""start：文档-主题分布矩阵，使用DocumentTopicMatrix[][]存储"""
DocumentTopicMatrix = np.zeros([4151, 20])#存储文档-主题分布
#存储文档-主题分布矩阵

for id in range(len(corpus)):
    doc_top = lda.get_document_topics(corpus[id], per_word_topics=False)
    m = 0
    for x in doc_top:
        DocumentTopicMatrix[id][m] = x[1]
        m += 1

#保存文档-主题分布矩阵
np.savetxt("data\DocumentTopicMatrix.txt",DocumentTopicMatrix)
print("文档-主题分布矩阵保存成功")
```

```python
# 主题确定
max_index = np.argmax(DocumentTopicMatrix,axis=1)
print(max_index)

f = codecs.open("./data/index.txt", "w", "utf-8")    # 打开文件以便写入中文
for i in max_index:  # content into txt
        print(i, file = f)
f.close()  #  关闭文件

```

```python
cnt=[]
for j in range(20):
    cont=[]
    index = 0
    num = 0 #主题大小元素
    for i in max_index:
        if i==j:
            cont.append(df.loc[index,"zh_titles"])
            num = num+1
        index = index+1
    cnt.append(num)
    # print(cont)
    # 存储不同类别
    path = "./data/content/content"+str(j+1).zfill(2)+".txt"
    f = codecs.open(path, "w", "utf-8")
    for i in cont:  # content into txt
        print(i, file = f)
    f.close()  #  关闭文件
```
